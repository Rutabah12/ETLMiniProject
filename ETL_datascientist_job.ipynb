```python
import pandas as pd
```

```python
from splinter import Browser
from bs4 import BeautifulSoup
import time
```

```python
executable_path = {"executable_path": "/usr/local/bin/chromedriver"}
browser = Browser("chrome", **executable_path, headless= False)
```

```python
url = "https://www.indeed.com/jobs?q=data+scientist&l="
browser.visit(url)
time.sleep(1)
```

```python
job_data = []
counter=0
scraping = True
while scraping == True:
    time.sleep(1)
    counter = counter+1
    print(f"COUNTER: {counter}")
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    jobs = soup.find_all("div", class_ = "jobsearch-SerpJobCard")

    for job in jobs:
        job_info = {}
        job_info["title"] = job.find("div", class_="title").a["title"]
        try:
            job_info["salary"] = job.find("div", class_ = "salarySnippet").text
        except:
            job_info["salary"] = "Unknown"
        try:
            job_info["location"] = job.find("div", class_="recJobLoc")["data-rc-loc"]
        except:
            job_info["location"] = "Unknown"
        try:
            job_info["company"] = job.find("span", class_="company").text
        except:
            job_info["company"] = "Unknown"
        job_data.append(job_info)
    try:
        print(browser.find_by_css(".np").last.text.lower())
        if "next" in browser.find_by_css(".np").last.text.lower():
            print(browser.find_by_css(".np").last.text.lower())
            browser.find_by_css(".np").last.click()
            time.sleep(1)
            print("next clicked")
        else:
            browser.find_by_css("#popover-x").click()
            scraping = False
    except:
        try:
            if browser.is_element_present_by_id("popover-x"):
                browser.find_by_css("#popover-x").click()
            else:
                scraping=False
        except:
             pass
    
        
print(job_data)
```

```python
data_scientist_job_df = pd.DataFrame(job_data)
data_scientist_job_df.head()
```

```python
data_scientist_job_df.to_csv("datascientistjobnew.csv", index=False)
```

```python
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import Column, Integer, String, Float
```

```python
# https://docs.sqlalchemy.org/en/13/dialects/postgresql.html#module-sqlalchemy.dialects.postgresql.psycopg2
# postgresql+psycopg2://user:password@host:port/dbname[?key=value&key=value...]

user = "postgres"
password = "changeme"
host = "localhost"
port = "5432" # 5432 for everyone else
db = "datajobsFort500_db"
uri = f"postgresql+psycopg2://{user}:{password}@{host}:{port}/{db}"
uri
```

```python
Base = declarative_base()
class User(Base):
    __tablename__ = "datajobs_fort500"
    job_id = Column(Integer, primary_key=True)
    job_title = Column(String(255))
    company = Column(String(255))
    job_location = Column(String(255))
    salary = Column(String(255))
    company_rank = Column(Integer)
    employees = Column(Float)
    revenues = Column(Float)
```

```python
# Create the connection engine
engine = create_engine(uri)
conn = engine.connect()
```

```python
# Base.metadata.drop_all(conn)
Base.metadata.create_all(conn)
```

```python
data_jobs_fort500_df.to_sql(User.__tablename__, conn, index=False, if_exists="append")
```
